```python
sns.set_style("whitegrid")
plt.figure(figsize=(10,6))
total_revenue = data.groupby("StockCode")["UnitPrice"].sum().sort_values(ascending=False).head(10)

total_revenue.plot(kind="bar")
plt.title("Total Revenue (UnitPrice)")
plt.xlabel("StockCode")
plt.ylabel(" sum of Revenue")
plt.show()
```

```python
sns.set_style("whitegrid")
plt.Figure(figsize=(14,10))

sales_tared=data.groupby("InvoiceDate")["UnitPrice"].sum().sort_values(ascending=False).head(10).plot(kind="line")
plt.title("sales tred over TIME")
plt.xlabel("date and time")
plt.ylabel("sales revenue over time")
plt.show()
```

```python
sns.set_style("darkgrid")
plt.Figure(figsize=(14,7))

sales_by_country=data.groupby("Country")["UnitPrice"].sum().sort_values(ascending=False).head(10)
sales_by_country.plot(kind="bar")
plt.title("sales by country")
plt.xlabel("Country")
plt.ylabel('UNIT PRICE')
plt.xticks(rotation=90)
plt.show()
```
